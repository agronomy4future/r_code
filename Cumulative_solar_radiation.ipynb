# Required packages

```r
library(readr)
library(ggplot2)
```

# Data upload

```r
github="https://raw.githubusercontent.com/agronomy4future/raw_data_practice/main/cumulative_solar_radiation.csv"
dataA=data.frame(read_csv(url(github), show_col_types=FALSE))
```

# Cumulative graph

```r
ggplot(data=dataA, aes(x=Days_from_sowing, y=Cumulative_PAR, color=Direction))+
  geom_line(aes(color=Direction), size=0.5) +
  geom_line(aes(color=Direction, linetype=Direction), size=0.5) +
  #
  geom_vline(xintercept= 2, linetype="dashed", color="grey") +
  geom_vline(xintercept= 32, linetype="dashed", color="grey") +
  geom_vline(xintercept= 63, linetype="dashed", color="grey") +
  geom_vline(xintercept= 94, linetype="dashed", color="grey") +
  geom_vline(xintercept= 124, linetype="dashed", color="grey") +
  #
  annotate("text", label="June", x=2, y=10000, angle=90,
           hjust=0, vjust=1.5, size=4) +
  annotate("text", label="July", x=32, y=11300, angle=90,
           hjust=0, vjust=1.5, size=4) +
  annotate("text", label="Aug", x=63, y=11300, angle=90,
           hjust=0, vjust=1.5, size=4) +
  annotate("text", label="Sep", x=94, y=11300, angle=90,
           hjust=0, vjust=1.5, size=4) +
  annotate("text", label="Oct", x=124, y=11300, angle=90,
           hjust=0, vjust=1.5, size=4) +
  annotate("text", label=paste("B)"), x=2, y=12000, size=7, family="serif",
           fontface="plain", color= "black") +
  geom_text(aes(family="serif",fontface=6), x=135, y=1600,
            label="(- 36.9%)", size=4, col="black") +
  geom_text(aes(family="serif",fontface=6), x=135, y=800,
            label="(- 18.4%)", size=4, col="black") +
  geom_text(aes(family="serif",fontface=6), x=135, y=0,
            label="(- 40.2%)", size=4, col="black") +
  scale_color_manual(values=c("red","black","blue","orange"))+
  scale_x_continuous(breaks=seq(0,147,21), limits=c(0,147)) +
  scale_y_continuous(breaks=seq(0,12000,2000), limits=c(0,12000)) +
  labs(x="Days after planting", y=bquote("Cumulative PAR (mol "~m^-2*")")) +
  theme_classic(base_size=18, base_family="serif")+
  #guides(color = guide_legend(keyheight=unit(0.5,"cm"), keywidth=unit(0.2, "cm"))) +
  theme(legend.position=c(0.7,0.15),
        legend.title=element_blank(),
        legend.key=element_rect(color=alpha(0.1), fill=alpha(0.1)),
        legend.text=element_text(family="serif", face="plain",
                                 size=13, color= "Black"),
        legend.background=element_rect(alpha(0.1)),
        axis.line=element_line(linewidth=0.5, colour="black"))
```
