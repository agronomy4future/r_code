```r
# to generate data
Genotype= c("A","B")
Yield= c(100,120)
dataA= data.frame(Genotype,Yield)

# to draw a graph
library(ggplot2)
ggplot(data=dataA, aes(x=Genotype, y=Yield)) +
  geom_bar(stat="identity",position="dodge", width=0.8,
           color="black", size=1.5) +
  scale_y_continuous(breaks = seq(0,150,30), limits = c(0,150)) +
  labs(x="Genotype", y="Yied") +
  theme_grey(base_size=18, base_family="serif")+
  theme(axis.line=element_line(linewidth=0.5, colour="black"))
```

# 1) How to add superscripts to the axis titles?

```r
ggplot(data=dataA, aes(x=Genotype, y=Yield)) +
  geom_bar(stat="identity",position="dodge", width=0.8,
           color="black", size=1.5) +
  scale_y_continuous(breaks = seq(0,150,30), limits = c(0,150)) +
 #code 1
  #labs(x="Genotype", y=bquote("Yield (ton ha"^"-1"*")")) +
 #code 2
  labs(x="Genotype", y=bquote("Yield (ton"~ha^-1*")")) +
  theme_grey(base_size=18, base_family="serif")+
  theme(axis.line=element_line(linewidth=0.5, colour="black"))
```

# 2) How to add subscripts to the axis titles?

```r
ggplot(data=dataA, aes(x=Genotype, y=Yield)) +
  geom_bar(stat="identity",position="dodge", width=0.8,
           color="black", size=1.5) +
  scale_y_continuous(breaks = seq(0,150,30), limits = c(0,150)) +
  labs(x=bquote("Genotype"["TM"]), y=bquote("Yield (ton"~ha^-1*")")) +
  theme_grey(base_size=18, base_family="serif")+
  theme(axis.line=element_line(linewidth=0.5, colour="black"))
```

# 3) How to add both superscript and subscripts to the axis titles?

```r
ggplot(data=dataA, aes(x=Genotype, y=Yield)) +
  geom_bar(stat="identity",position="dodge", width=0.8,
           color="black", size=1.5) +
  scale_y_continuous(breaks = seq(0,150,30), limits = c(0,150)) +
  labs(x=bquote("Genotype"["TM"]),
       y=bquote("CO"["2"]~"assimilation rate"~(mu~mol ~CO[2]~ m^-2~s^-1))) +
  theme_grey(base_size=18, base_family="serif")+
  theme(axis.line=element_line(linewidth=0.5, colour="black"))
```

# 4) How to add superscript and subscripts to the axis text?

```r
ggplot(data=dataA, aes(x=Genotype, y=Yield)) +
  geom_bar(stat="identity",position="dodge", width=0.8,
           color="black", size=1.5) +
  scale_y_continuous(breaks = seq(0,150,30), limits = c(0,150)) +
  scale_x_discrete(limits=c("A","B"),
                   label= c(bquote("A"^"Europe"),
                            bquote("B"["Asia"]))) +
  labs(x=bquote("Genotype"["TM"]),
       y=bquote("CO"["2"]~"assimilation rate"~(mu~mol ~CO[2]~ m^-2~s^-1))) +
  theme_grey(base_size=18, base_family="serif")+
  theme(axis.line=element_line(linewidth=0.5, colour="black"))
```

# Full code

```r
Genotype= c("A","B")
Yield= c(100,120)
dataA= data.frame(Genotype,Yield)

library(ggplot2)
ggplot(data=dataA, aes(x=Genotype, y=Yield)) +
  geom_bar(stat="identity",position="dodge", width=0.8,
           color="black", size=1.5) +
  scale_y_continuous(breaks = seq(0,150,30), limits = c(0,150)) +
  scale_x_discrete(limits=c("A","B"),
                   label= c(bquote("A"^"Europe"), bquote("B"["Asia"])))  +
  labs(x=bquote("Genotype"["TM"]),
       y=bquote("CO"["2"]~"assimilation rate"~(mu~mol ~CO[2]~ m^-2~s^-1))) +
  theme_grey(base_size=18, base_family="serif")+
  theme(axis.line=element_line(linewidth=0.5, colour="black"))
```
